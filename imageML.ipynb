```python
#preprocessing
#1. resize
#2. flatten

import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread #you can use opencv library as an alternative
from skimage.transform import resize #images must all be in one specific size

target =[]
images=[]
flat_data=[]


DATADIR = '/content/images'
CATEGORIES = ['bulldog','doberman dog','german shepherd dog']

#iterate through them
for category in CATEGORIES:
  class_num = CATEGORIES.index(category)#label encoding the values to reduce time to go through all images
  path = os.path.join(DATADIR,category)#create path to use all the images
  for img in os.listdir(path):
    if os.path.isfile(os.path.join(path, img)):
      img_array = imread(os.path.join(path,img))
      print(img_array.shape)
      img_resized=resize(img_array,(150,150,3))
      flat_data.append(img_resized.flatten())
      images.append(img_resized)
      target.append(class_num)

flat_data= np.array(flat_data)
target= np.array(target)
images= np.array(images)
```

```python
target
```

```python
unique,count= np.unique(target,return_counts=True)
plt.bar(CATEGORIES,count)
```

```python
#split data into training and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(flat_data,target,test_size=0.3,random_state=100)
```

```python
#using support vector m
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid=[
    {'C':[1,10,100,1000],'kernel':['linear']},
    {'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']},
]
svc=svm.SVC(probability=True)#tells you probability of an image being in a class
clf=GridSearchCV(svc,param_grid)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
import pickle
pickle.dump(clf,open('img_model.p','wb'))
```
